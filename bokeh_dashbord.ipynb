```python

from bokeh.plotting import figure, output_file, show
from bokeh.tile_providers import STAMEN_TERRAIN,OSM,ESRI_IMAGERY, get_provider
from bokeh.io import output_notebook
import pandas as pd
```

```python
output_notebook()
```

```python
df_combined = pd.read_pickle('wgms_combined')
```

```python
output_file("tile.html")

tile_provider = get_provider(ESRI_IMAGERY)
# range bounds supplied in web mercator coordinates
p = figure(x_range=(-20000000, 20000000), y_range=(-1000000, 7000000),
           x_axis_type="mercator", y_axis_type="mercator")
p.add_tile(tile_provider)
p.circle(x=df_combined['merX'], y=df_combined['merY'])

show(p)
```
